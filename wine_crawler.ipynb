```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
from meteostat import Stations, Monthly

file_path = '/content/drive/MyDrive/Colab_Notebooks/Data_Mining_Project/'
red_path = file_path + 'Red_Coordi.csv'  # 將路徑替換為您的檔案路徑
white_path = file_path + 'White_Coordi.csv'  # 將路徑替換為您的檔案路徑
Red = pd.read_csv(red_path)
White = pd.read_csv(white_path)

for file_name, name in [(Red, 'Red'), (White, 'White')]:

  # 移除 'N.V.' 資料
  file_name = file_name[file_name['Year'] != 'N.V.']
  #Red['Year'] = Red['Year'].astype(str)
  # 使用 pd.to_datetime 將年份字串轉換為日期時間物件
  file_name['Year'] = pd.to_datetime(file_name['Year'], format='%Y')

  # 定義月份範圍
  seasons = {
      '3-5': [3, 4, 5],  # 春季
      '6-8': [6, 7, 8],  # 夏季
      '9-11': [9, 10, 11]  # 秋季
  }

  # 定義要新增的欄位
  columns_to_add = ['tavg', 'tmin', 'tmax', 'prcp', 'tsun']

  # 動態建立新的欄位名稱
  for season in seasons.keys():
      for col in columns_to_add:
          Red[f'{season}_{col}'] = None

  #print(Red.head())

  def get_weather_data(lat, lng, year):
  #for lat, lng, year in zip(Red['lat'], Red['lng'], Red['Year']):
    stations = Stations().nearby(lat, lng)
    station = stations.fetch(1)

    if station.empty:
      #print('\nstation_id=', station_id)
      return None

    station_id = station['wmo'].iloc[0]  # 取得 ID 值
    print(station_id)

    # Set time period
    start = datetime(year.year-1, 3, 1)
    end = datetime(year.year-1, 11, 30)

    # Get Monthly data
    data = Monthly(station_id, start, end).fetch()

    # 將 time 列轉換為 datetime 格式，方便篩選月份
    data.index = pd.to_datetime(data.index, format='%Y-%m-%d')
    return data


  def process_wine(row):
    print(row.name)
    lat, lng, year = row['lat'], row['lng'], row['Year']
    data = get_weather_data(lat, lng, year)


    if data is not None:
        # 計算每個季節的平均值並更新到 Red
        for season, months in seasons.items():
            season_data = data[data.index.month.isin(months)].mean()
            #print(season_data)
            for col in columns_to_add:
                row[f'{season}_{col}'] = season_data[col]

    return row

  # 使用 apply 函數處理每一行資料
  file_name = file_name.apply(process_wine, axis=1)
  # 顯示結果
  print(file_name.head())
  file_name.to_csv(file_path + name + '_Weather.csv', index=False)

'''
Column	Description
station	The Meteostat ID of the weather station
time	The month, represented as a date
tavg	The average air temperature in °C
tmin	The minimum air temperature in °C
tmax	The maximum air temperature in °C
prcp	The monthly precipitation total in mm
wspd	The average wind speed in km/h
pres	The average sea-level air pressure in hPa
tsun	The monthly sunshine total in minutes (m)
'''
```

```python
!pip install openmeteo-requests
!pip install requests-cache retry-requests numpy pandas

import openmeteo_requests
from datetime import datetime
import requests_cache
import pandas as pd
from retry_requests import retry
import time
import os

def openmeteo(lat, lng, year):
  time.sleep(1)
  # Setup the Open-Meteo API client with cache and retry on error
  cache_session = requests_cache.CachedSession('.cache', expire_after = -1)
  retry_session = retry(cache_session, retries = 5, backoff_factor = 0.2)
  openmeteo = openmeteo_requests.Client(session = retry_session)

  # Make sure all required weather variables are listed here
  # The order of variables in hourly or daily is important to assign them correctly below
  url = "https://archive-api.open-meteo.com/v1/archive"
  params = {
    "latitude": lat,
    "longitude": lng,
    "start_date": str(int(year)-1) + "-01-01",
    "end_date": str(int(year)-1) + "-12-31",
    "daily": ["temperature_2m_max", "temperature_2m_min", "temperature_2m_mean", "sunshine_duration", "precipitation_sum"]
  }
  responses = openmeteo.weather_api(url, params=params)

  # Process first location. Add a for-loop for multiple locations or weather models
  response = responses[0]
  #print(f"Coordinates {response.Latitude()}°N {response.Longitude()}°E")
  #print(f"Elevation {response.Elevation()} m asl")
  #print(f"Timezone {response.Timezone()} {response.TimezoneAbbreviation()}")
  #print(f"Timezone difference to GMT+0 {response.UtcOffsetSeconds()} s")

  # Process daily data. The order of variables needs to be the same as requested.
  daily = response.Daily()
  daily_temperature_2m_max = daily.Variables(0).ValuesAsNumpy()
  daily_temperature_2m_min = daily.Variables(1).ValuesAsNumpy()
  daily_temperature_2m_mean = daily.Variables(2).ValuesAsNumpy()
  daily_sunshine_duration = daily.Variables(3).ValuesAsNumpy()
  daily_precipitation_sum = daily.Variables(4).ValuesAsNumpy()

  daily_data = {"date": pd.date_range(
    start = pd.to_datetime(daily.Time(), unit = "s", utc = True),
    end = pd.to_datetime(daily.TimeEnd(), unit = "s", utc = True),
    freq = pd.Timedelta(seconds = daily.Interval()),
    inclusive = "left"
  )}
  daily_data["tmax_avg"] = daily_temperature_2m_max
  daily_data["tmin_avg"] = daily_temperature_2m_min
  daily_data["tmean_avg"] = daily_temperature_2m_mean
  daily_data["sunshine_avg"] = daily_sunshine_duration
  daily_data["prcp_avg"] = daily_precipitation_sum

  daily_dataframe = pd.DataFrame(data = daily_data)
  #print(daily_dataframe)
  # 確保 'date' 列是 datetime 類型，並且設定為索引
  daily_dataframe['date'] = pd.to_datetime(daily_dataframe['date'])
  daily_dataframe.set_index('date', inplace=True)

  # 使用 resample('M') 來重新取樣為每月，然後對各個變量取平均值
  monthly_avg = daily_dataframe.resample('ME').mean()

  # 打印結果
  return monthly_avg



file_path = '/content/drive/MyDrive/Colab_Notebooks/Data_Mining_Project/'
red_path = file_path + 'Rose_Coordi.csv'  # 將路徑替換為您的檔案路徑
white_path = file_path + 'Sparkling_Coordi.csv'  # 將路徑替換為您的檔案路徑
Red = pd.read_csv(red_path)
White = pd.read_csv(white_path)

#print(Red.head())

for file_name, name in [(Red, 'Rose'), (White, 'Sparkling')]:

  # 移除 'N.V.' 資料
  file_name = file_name[file_name['Year'] != 'N.V.']

  # 讀取之前保存的 CSV（如果存在）
  try:
      previous_data = pd.read_csv(file_path + name + '_New_Weather.csv')
      processed_rows = len(previous_data)  # 獲取已處理的行數
      print(f"已處理的行數: {processed_rows}")
  except FileNotFoundError:
      processed_rows = 0  # 如果檔案不存在，從頭開始處理
      print("檔案不存在，從頭開始處理資料。")

  # 僅處理未處理的資料
  file_name = file_name.iloc[processed_rows:]  # 獲取尚未處理的資料

  months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

  # 定義要新增的欄位
  columns_to_add = ['tavg', 'tmin', 'tmax', 'prcp', 'tsun']

  # 動態建立新的欄位名稱
  for month in months:
      for col in columns_to_add:
          Red[f'{month}_{col}'] = None

  # 新增欄位
  def process_wine(row):
    print(row.name+1)
    lat, lng, year = row['lat'], row['lng'], row['Year']
    data = openmeteo(lat, lng, year)
    #print(data)
    #input()
    if data is not None:
        # 確保 'date' 列是 datetime 類型，並且設定為索引
        #data['date'] = pd.to_datetime(data['date'])
        #data.set_index('date', inplace=True)

        # 將每個月的平均值插入到對應的 Red 資料框中
        for i, month in enumerate(months):
            if i + 1 in data.index.month:
                row[f'{month}_tavg'] = data.loc[data.index.month == (i + 1), 'tmean_avg'].values[0]
                row[f'{month}_tmin'] = data.loc[data.index.month == (i + 1), 'tmin_avg'].values[0]
                row[f'{month}_tmax'] = data.loc[data.index.month == (i + 1), 'tmax_avg'].values[0]
                row[f'{month}_prcp'] = data.loc[data.index.month == (i + 1), 'prcp_avg'].values[0]
                row[f'{month}_tsun'] = data.loc[data.index.month == (i + 1), 'sunshine_avg'].values[0]

    # 檢查檔案是否存在，如果不存在，寫入表頭
    if not os.path.isfile(file_path + name + '_New_Weather.csv'):
        row.to_frame().T.to_csv(file_path + name + '_New_Weather.csv', index=False, mode='w', header=True)
    else:
        # 使用追加模式將每一筆處理完的資料寫入 CSV，不寫入表頭
        row.to_frame().T.to_csv(file_path + name + '_New_Weather.csv', index=False, mode='a', header=False)

    return row


  # 使用 apply 函數處理每一行資料
  if not file_name.empty:
    for idx, row in file_name.iterrows():
      process_wine(row)  # 逐筆處理
    print('所有資料已處理並儲存。')
  else:
    print(f"所有資料已經處理過了，不需要更新。")
```
